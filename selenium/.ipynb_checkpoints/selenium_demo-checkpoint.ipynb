```python
driver = webdriver.Chrome()
driver.maximize_window()

url = 'https://www.google.com'
driver.get(url)
time.sleep(1)

#location the search bar
search_bar_xpath = '//*[@id="APjFqb"]'
search_bar = driver.find_element(by=By.XPATH,value=search_bar_xpath)

#for input use send_keys functions
search_bar.send_keys("machine learning")
time.sleep(1)

#to clear the input values
# search_bar.clear()
# time.sleep(1)

#clicking keys
search_bar.send_keys(Keys.ENTER)
time.sleep(2)

# #clicking action or 
# link_xpath = '//*[@id="rso"]/div[2]/div/div/div/div[1]/div/div/span/a/h3'

# link = driver.find_element(By.XPATH,link_xpath)

# link.click()
driver.quit()
```

```python
driver = webdriver.Chrome()
driver.maximize_window()

url = 'https://github.com/login'
driver.get(url)
time.sleep(1)

#username field
username_field = driver.find_element(By.ID,'login_field')
username_field.send_keys('awaz')

#password field
password_field = driver.find_element(By.ID, "password")
password_field.send_keys('pass123')
time.sleep(1)

#submit button 
submit_button = driver.find_element(By.XPATH,'/html/body/div[1]/div[3]/main/div/div[2]/form/div[3]/input')
submit_button.click()

driver.quit()
```

working with dropdowns

```python
from selenium import webdriver
import time
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.select import Select
```

```python
driver = webdriver.Chrome()
driver.maximize_window()
url = 'https://www.miniclip.com/careers/vacancies'
driver.get(url)
time.sleep(2)

departments_field = driver.find_element(By.XPATH,'//*[@id="__layout"]/div/div/section[2]/div/fieldset[3]/select')
departments_dropdown = Select(departments_field)

#departments_dropdown.select_by_index(5)
#time.sleep(1)

departments_dropdown.select_by_visible_text('Art')
driver.quit()

#multi select in dropdown box
'''cars_ms = Select(cars_element)
cars_ms.select_by_index(1)
time.sleep(2)

cars_ms.select_by_visible_text('Opel')
time.sleep(2)

cars_ms.select_by_visible_text('Audi')
time.sleep(2)

#to remove the select 
cars_ms.deselect_by_index(2)
time.sleep(2)

cars_ms.deselect_all() #remove all the select option from the list of dropdown
time.sleep(2)
driver.quit()
'''
```

```python
#Scrolling by elements
driver = webdriver.Chrome()
driver.maximize_window()
url  ='https://en.wikipedia.org/wiki/Machine_learning'
driver.get(url)
time.sleep(2)

ai_xpath = '//*[@id="toc-Artificial_intelligence"]'
ai_subtopic = driver.find_element(By.XPATH,ai_xpath)
driver.execute_script('arguments[0].scrollIntoView();',ai_subtopic)

driver.execute_script("window.scrollBy(0,1000);")
time.sleep(2)
driver.execute_script("window.scrollBy(0,-1000);")
time.sleep(2)

driver.execute_script("window.scrollBy(-1000,0);")
time.sleep(2)
```
